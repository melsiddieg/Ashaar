```python
import inspect, pathlib, shutil, poetry_diacritizer

# 1) Where did pip install the wheel?
pkg_dir = pathlib.Path(inspect.getfile(poetry_diacritizer)).parent
print("• Wheel directory :", pkg_dir)

# 2) Where is the repo’s config folder (relative to the notebook cwd)?
repo_cfg_dir = pathlib.Path("poetry_diacritizer") / "config"
if not repo_cfg_dir.is_dir():
    raise FileNotFoundError(f"Config folder not found at {repo_cfg_dir.resolve()} —"
                            " are you running this cell from the project root?")

print("• Repo config dir :", repo_cfg_dir.resolve())

# 3) Copy everything over (keep existing files if they’re already there)
shutil.copytree(repo_cfg_dir, pkg_dir / "config", dirs_exist_ok=True)
print("✓ YAMLs copied:")
for f in (pkg_dir / "config").iterdir():
    print("  -", f.name)
```
